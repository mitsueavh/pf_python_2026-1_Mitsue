```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
```

```python
f="../data/001_raw/20241210/Esolmet_CR6_IP_Table2_10min.dat"
esol=pd.read_csv(f, sep=",",low_memory=False,skiprows=[0,2,3],index_col=0,parse_dates=True)
esol.info()
```

```python
esol_limpio = esol.sort_index() #Acomodar el indice
esol_limpio = esol_limpio["2023-07-01":"2024-06-30"] #Cortar el df a 12 meses
esol_limpio["I_glo_Avg"] = pd.to_numeric(esol_limpio['I_glo_Avg'], errors='coerce') #Convertir Ig e Id a float
esol_limpio["I_dir_Avg"] = pd.to_numeric(esol_limpio['I_dir_Avg'], errors='coerce') #Convertir Ig e Id a float
esol_limpio = esol_limpio[['I_glo_Avg', 'I_dir_Avg', 'I_dif_Avg','WS_ms_Avg', 'WindDir', 'AirTC_Avg', 'RH', 'CS106_PB_Avg']] #Seleccionar solo las columnas de interes
esol_limpio.info()
```

```python
msno.matrix(esol_limpio)
#No se identificaron datos faltantes
esol_limpio.isna().resample("ME").sum()
```

```python
esol_limpio.to_csv("../data/002_processed/2024.csv")
esol_limpio.to_parquet("../data/002_processed/2024.parquet")
```
